```python
import pandas as pd
from IPython.display import display
df=pd.read_csv(r"C:\Users\Rohini jadhav\Downloads\intense_stock_data.csv")
df["Volatility"]=df["High"]-df["Low"]
volatility=df["Volatility"]
print(f"volatility : {volatility}")

# to find the stock with high volatility days(volatility>100)
def classify_volatility(v):
    if v > 100:
        return "High volatility"
    else: 
        return"Low volatility"
        
df["Volatility Performance"]=df["Volatility"].apply(classify_volatility)
counts= df["Volatility Performance"].value_counts()
print (f"\nDay type count: {counts}")

high_volatility_stock=df[df["Volatility Performance"]=="High volatility"]
grouped_data=high_volatility_stock.groupby("Stock").size()

all_stock_volatility= grouped_data.idxmax()
most_high_volatile_stock=grouped_data.max()

top_five=df.sort_values(by="Volatility",ascending=False).head(5)

print(f"\ntop five stokcs with   high volatility days: \n{top_five[['Date','Stock','Volatility']]}")
print(f"\n stocks with high volatility days:\n{grouped_data}")
print(f"\nstock with most volatility days :\n{most_high_volatile_stock} : {all_stock_volatility}")

# creating extra rows of code cus i am yet begginer  using same term by modifying it 
# and applying it,thats still hard, i get confused , so i am taking long route , but 
#i will be there eventually  with practice .

def volatility_type(Volatility):
    if Volatility> 100:
        return"gain"
    elif Volatility<100:
        return"loss"
    else: 
        return"no change"
df["Volatility"]=df["High"]-df["Low"]       
df["volatility type"]=df["Volatility"].apply(volatility_type)
avg_volatility=df.groupby("volatility type")["Volatility"].mean().round(2)
print(f"\naverage volatility compare :\n{avg_volatility}")
display(df)
```
