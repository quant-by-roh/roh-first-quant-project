```python
import pandas as pd 
data={
     "Stock": ["TCS", "Infosys", "HDFC", "Wipro"],
    "Buy Price": [3200, 1500, 1725, 410],
    "Current Price": [3450, 1420, 1800, 390],
}
df=pd.DataFrame(data)
from IPython.display import display


df["Profit"]=df["Buy Price"]-df["Current Price"]

df["Return %"]=((df["Current Price"]-df["Buy Price"])/df["Buy Price"]*100)
df["Return %"] = df["Return %"].round(2)
def rate_stock(r):
  if r>10:
    return"strong"
  elif 10>=r>=0:
    return"moderate"
  else :
     return"weak"
df["Rating"]=df["Return %"].apply(rate_stock)
display(df)
# first i displayed after all three df and got three different tables first with only 
#given set od data and profit,then second with return and third with rating,
#it wasnot looking all that great and clean i only displayed or printed once at end 
#to put it all together. 
    
```
