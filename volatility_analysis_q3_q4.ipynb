```python
import pandas as pd 
from IPython.display import display
df=pd.read_csv(r"C:\Users\Rohini jadhav\Downloads\intense_stock_data.csv")
df["Volatility"]=df["High"]-df["Low"]
def classify_volatility(v):
    if v > 100:
        return"high volatility"
    elif 100>= v >= 50:
        return "medium volatility"
    elif 50 >v:
        return"low volatility"
df["Performance"]= df["Volatility"].apply(classify_volatility)
performance_counts= df["Performance"].value_counts()
high_volatility_stock=df[df["Performance"]=="high volatility"]
grouped_data= high_volatility_stock.groupby("Stock").size()
high_volatile_stock=grouped_data.max()
high_volatile_stocks=grouped_data.idxmax()
print(f"\nstock with most high volatility_days : {high_volatile_stock} : {high_volatile_stocks}")
# here i am going to count total number of high volatility stocks and display it but 
#since i have already count performance count where it mentions the high volatility count
# i added  the high_volatile_stock
print(f"\nperformance type count : {performance_counts}")

# most stable stock, lower average volatility

average_volatility_per_stock=df.groupby("Stock")["Volatility"].mean()
most_stable_stock= average_volatility_per_stock.idxmin()
most_stable_volatility=average_volatility_per_stock.min()

print(f"\nmost stable stock : \n{most_stable_stock} \nwith average voaltility:\n{most_stable_volatility}")


display(df)
```
