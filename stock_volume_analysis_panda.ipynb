```python
import pandas as pd 
from IPython.display import display
df=pd.read_csv(r"C:\Users\Rohini jadhav\Downloads\intense_stock_data.csv")
df["Change"]=df["Close"]-df["Open"]
df["Performance"]=df["Change"].apply(
      lambda x: "Gain" if x > 0 else ("Loss" if x < 0 else "No Change")
  )
                                      
Stock_Categorey_Count=df["Performance"].value_counts()
print(f"count of performance : {Stock_Categorey_Count}")
stock_categorey_count=df["Change"].value_counts().head(5)
print(f"\ncount of first 5 rows: {stock_categorey_count}")
def strength_stock(Change):    # here we custom func we cant use word according gto our
    if Change>50:              # conveinience we have to use change so the conditions
        return"strong gain"    # will run through that func 
    elif 50 >= Change >0:
        return"mild gain"
    elif 0>Change >= -50:
        return"mild loss"
    elif  Change<-50:
        return"strong loss"
df["strength"]=df["Change"].apply(strength_stock)
strength_count=df["strength"].value_counts()
print(f"\nstrength count of stocks : {strength_count}")
strength_rows=df["strength"].sort_values(ascending=False)
print(f"\nstrong gain rows = {strength_rows}")
high_thresh=df["Volume"].quantile(0.75)
low_thresh=df["Volume"].quantile(0.25)

def classify_volume(v):
    if v>high_thresh:
        return"high volume"
    elif v<= low_thresh:
        return"low volume"
    else :
        return"medium volume"
df["volume categorey"]=df["Volume"].apply(classify_volume)
volume_performance=df.groupby(["volume categorey","Performance"]).size()
print(f"\nVolume categorey vs performance:\n{volume_performance}")

top_gain_volume_days=df.sort_values(by="Volume",ascending=False).head(5)
print(f"\n top 5 highest volume days:\n" f"{top_gain_volume_days[['Date','Stock','Volume','Performance']]}")

#Find the top 3 lowest volume days where strength = strong gain.

strong_gain_days=df[df["strength"]=="strong gain"]
Top_three_low_volume_days=strong_gain_days.sort_values(by="Volume",ascending=True).head(3)
print(f"\n top 3 low volume days : \n{Top_three_low_volume_days[['Date','Stock','Volume','strength']]}")

#which stock have the highest number of high volume days classified as Gain.

gain_days=df[df["Performance"]=="Gain"]
stock_volume_performance=gain_days[gain_days["volume categorey"]== "high volume"]
grouped_data=stock_volume_performance.groupby("Stock").size()

top_stock= grouped_data.idxmax()
top_count=grouped_data.max()
print(f"\ndays with high volume ,categarized as  gain \n{grouped_data}")
print(f"\nstock with most high volume days categorized as gain: \n{top_stock}: {top_count}")

#average change per volume categoreyave

df["change"]=df["Close"]-df["Open"]
average_volume_change=df.groupby("volume categorey")["change"].mean().round(2)
overall_average_volume_change=average_volume_change.mean().round(2)
print(f"\naverage change per volume :\n{average_volume_change}")
print(f"\noverall average change : \n{overall_average_volume_change}")


# creating column for volatility and finding out the volatility
df["Volatility"]=df["High"]-df["Low"]
volatility_per_stock=df.groupby("Stock")["Volatility"].mean().sort_values(ascending=False).round(2)
overall_volatility=volatility_per_stock.mean().round()
print(f"\nvolatility per stock :\n{volatility_per_stock}")
print(f"\noverall volatility : \n{overall_volatility}")

top_three_stock=df.sort_values(by="Volatility",ascending=False).head(3)
print(f"\ntop three volatile stocks :\n{top_three_stock[['Date','Stock','Volatility']]}")
display(df)
```
